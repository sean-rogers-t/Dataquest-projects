```python
sample_means=[]
parameter=wnba['PTS'].mean()
for i in range(100):
    samp=wnba['PTS'].sample(10,random_state=i)
    sample_mean_pts=samp.mean()
    sample_means.append(sample_mean_pts)
fig,ax=plt.subplots()
ax.scatter(range(1,101),sample_means)
ax.set_xlabel("Sample Number")
ax.set_ylabel("Sample Mean")
plt.axhline(parameter)
plt.show()

```
